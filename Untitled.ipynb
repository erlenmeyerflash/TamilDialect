```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("xlm-roberta-base")

pairs = [
    ("செய்கிறேன்", "செய்றேன்"),      # literary vs spoken
    ("வைத்தியசாலை", "மருத்துவமனை"),  # SL Tamil vs Indian Tamil for hospital
    ("கதைக்கிறேன்", "பேசுகிறேன்"),   # SL Tamil vs Indian Tamil for speaking
    ("உந்துருளி", "மிதிவண்டி"),        # SL Tamil vs Indian Tamil for bicycle
]

for sl_word, indian_word in pairs:
    sl_tokens = tokenizer.tokenize(sl_word)
    in_tokens = tokenizer.tokenize(indian_word)
    print(f"SL: {sl_word} → {sl_tokens} ({len(sl_tokens)} tokens)")
    print(f"IN: {indian_word} → {in_tokens} ({len(in_tokens)} tokens)")
    print()
```
